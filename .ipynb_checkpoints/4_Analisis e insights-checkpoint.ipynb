### *¿Qué factores (a parte de la localización) determinan el precio del alquiler?*

En la preparacion de los datos habíamos discretizado varias variables para que nos permitiran analizar esto de una manera más sencilla.

Para ello aplicaremos la técnica de los "cubos" (cubos OLAP), que consiste en organizar la información en métricas (que se requieren analizar) y dimensiones (vista de análisis)

```python
df.filter(like='disc')
```

#### Creación minicubo

```python
# Paso 1: Seleccionar las métricas y las dimensiones
metricas = ['total_price', 'precio_compra']
dimensiones = ['bedrooms_disc', 'beds_disc', 'accommodates_disc', 'number_of_reviews_disc']

cubo_precio = df.loc[:, dimensiones + metricas] #seleccion de columnas
cubo_precio
```

```python
# Paso 2: Pasar de formato "ancho" (tabular) a formato "largo" (transaccional)

cubo_precio = cubo_precio.melt(id_vars=metricas)
cubo_precio
```

```python
# Paso 3: Agregar las métricas por "variable" y "value" con las funciones deseadas

cubo_precio = cubo_precio.groupby(['variable', 'value'])[['total_price','precio_compra']].agg('median')
cubo_precio
```

### Análisis de la influencia de la competencia sobre el precio del alquiler

En este análisis, exploraremos cómo el número de inmuebles disponibles por barrio (competencia) afecta el precio del alquiler (`total_price`).

En el análisis de ranking de distritos y barrios por precio medio de alquiler habíamos obtenido un listado de posibles barrios de interes que apriori pueden maximizar la relación coste-ingresos

```python
competencia_alquiler = df.groupby('neighbourhood').agg(
    num_imbuebles = ('id', 'count'),
    media_precio_alquiler = ('total_price', 'median')
).reset_index()
competencia_alquiler
```

```python
total_listings = competencia_alquiler.num_imbuebles.sum()
competencia_alquiler['porcentaje_competencia'] = 100*competencia_alquiler['num_imbuebles'] / total_listings
competencia_alquiler
```

```python
top_30_alquiler = competencia_alquiler.sort_values(by='porcentaje_competencia',ascending=False).head(30)
top_30_alquiler
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(top_30_alquiler['neighbourhood'],top_30_alquiler['porcentaje_competencia'])
ax.set_xticks(range(len(top_30_alquiler['neighbourhood'])))
ax.set_xticklabels(top_30_alquiler['neighbourhood'], rotation=90)
ax.set_title('Porcentaje de competencia vs Precio alquiler (Top 30 barrios por listings)')
ax.set_ylabel('Porcentaje de competencia (% sobre Total)')
ax.set_xlabel('Barrio')

ax2 = ax.twinx()
ax2.plot(top_30_alquiler['neighbourhood'],top_30_alquiler['media_precio_alquiler'], c='orange', marker='o')
ax2.set_ylabel('Precio medio alquiler (€)')

plt.grid(True, alpha=0.3);
```

#### 💡Insight 7
**No se aprecia una relación clara entre el precio del alquiler y el número de listings, no implica directamente precio de alquileres más bajos. Otros factores serán más influeyentes en el precio del alquiler**

### ¿Cómo influye la competencia (num inmuebles disponibles por barrio) sobre la ocupación?

```python
competencia_ocupacion = df.groupby('neighbourhood').agg(
                            num_inmuebles = ('id', 'count'),
                            ocupacion_media =('ocupacion', 'median')
                        ).reset_index()
competencia_ocupacion
```

```python
total_inmuebles = competencia_ocupacion['num_inmuebles'].sum()
total_inmuebles
```

```python
competencia_ocupacion['porcentaje_competencia'] = 100 * (competencia_ocupacion['num_inmuebles'] / total_inmuebles)
competencia_ocupacion
```

```python
top_competicion_ocupacion = competencia_ocupacion.sort_values(by = 'porcentaje_competencia', ascending=False).head(30)
top_competicion_ocupacion
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(top_competicion_ocupacion['neighbourhood'],top_competicion_ocupacion['porcentaje_competencia'])
ax.set_xticks(range(len(top_competicion_ocupacion['neighbourhood'])))
ax.set_xticklabels(top_competicion_ocupacion['neighbourhood'], rotation=90)
ax.set_title('Porcentaje de competencia vs Ocupación (Top 20 barrios por anuncios)')
ax.set_ylabel('Porcentaje de competencia (% sobre Total)')
ax.set_xlabel('Barrio')

ax2 = ax.twinx()
ax2.plot(top_competicion_ocupacion['neighbourhood'],top_competicion_ocupacion['ocupacion_media'], c='orange', marker='o')
ax2.set_ylabel('Ocupacion media (%)')

plt.grid(True, alpha=0.3);

```

##### 💡Insight 15:

**No existe una relación lineal clara entre el número de anuncios y la ocupación. Una alta competencia no implica directamente una menor ocupación**
**Pero si se observan barrios como Recoletos, Delicias, Cuatro Caminos, Pacífico, Puerta del Angel, Castellana, Arapiles y Almagro los cuales tiene un menor número de anuncios y su ocupación media se mantiene estable en ~75%**

Vamos a analizar visualemnte el insight sobre el distrio de San Blas.

```python
import folium
```

```python
# Incluimos las coordendas de inicio (Puerta del Sol) y el zoom
mapa = folium.Map(location=[40.4167278, -3.7033387],zoom_start=12)
```

```python
datos = df[df.distrito == 'San Blas - Canillejas'].copy()
datos
```

```python
mapa = folium.Map(location=[40.4167278, -3.7033387],zoom_start=12)

for piso in range(0,len(datos)):
   folium.Marker(
      location = [datos.iloc[piso]['latitude'], datos.iloc[piso]['longitude']],
      popup = datos.iloc[piso]['total_price'],
   ).add_to(mapa)

mapa
```

Realizamos un código de colores que nos indicara visualmente el rango de precio.

Para ello vamos a crear una nueva variable discretizada y cambiaremos el marcador a un círculo de colores.

```python
datos['precio_total_disc'] = pd.qcut(datos['total_price'], q = [0, .25, .5, .75, 1.], 
                              labels=['yellow', 'orange', 'blue', 'red'])
```

```python
mapa = folium.Map(location=[40.4167278, -3.7033387],zoom_start=12)

for piso in range(0,len(datos)):
   folium.CircleMarker(
      location = [datos.iloc[piso]['latitude'], datos.iloc[piso]['longitude']],
      popup = datos.iloc[piso]['total_price'],
      fill=True,
      color = datos.iloc[piso]['precio_total_disc'],
      fill_opacity=1,
      radius = 5
   ).add_to(mapa)


mapa.save("mapa_San_Blas.html")
mapa
```

##### 💡Insight 16
**Existen todavía muchos puntos amarillos y naranjas, lo cual puede indicar que todavía existe potencial para explotar más el factor de los eventos deportivos**
